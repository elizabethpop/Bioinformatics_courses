```python
#Function ReverseComplement(text) finds the reverse complement of a DNA string

def ReverseComplement(text):
    letters = { "A":"T", "C":"G", "G":"C", "T":"A"}
    seq_rev = ''
    for i in range(len(text)):
        seq_rev += letters[text[len(text) - i - 1]]
    return seq_rev
```

```python
#Function HammingDistance(text_1, text_2) calculates the Hamming distance (the number of mismatches
# between strings) between two strings of the same length.

def HammingDistance(text_1, text_2):
    text_spl_1 = [x for x in text_1]
    text_spl_2 = [x for x in text_2]
    hamming_dist = 0
    for i in range(len(text_1)):
        if text_spl_1[i] != text_spl_2[i]:
            hamming_dist += 1
    return(hamming_dist)
```

```python
#Function ApproximatePatternMatching(Pattern, Text, k) finds all approximate (with the HammingDistance <= k)
# occurrences of a pattern in a string.

def ApproximatePatternMatching(Pattern, Text, k):
    ind_few_mismatches = []
    text_list = []
    for i in range(len(Text)-len(Pattern)+1):
        text_list.append(Text[i:(i+len(Pattern))])
        for j in text_list:
            Spl_1 = [x for x in j]
            Spl_Pattern = [x for x in Pattern]
            hamming_dist = HammingDistance(Spl_1, Spl_Pattern)
        if hamming_dist <= k:
            ind_few_mismatches.append(i)
    return(ind_few_mismatches)
```

```python
#Function ApproximatePatternCount(Pattern, Text, d) computes the Hamming distance between Pattern and
# every k-mer substring of Text

def ApproximatePatternCount(Pattern, Text, d):
    count = 0
    text_list = []
    for i in range(len(Text)-len(Pattern)+1):
        text_list.append(Text[i:(i+len(Pattern))])
        for j in text_list:
            Spl_1 = [x for x in j]
            Spl_Pattern = [x for x in Pattern]
            hamming_dist = HammingDistance(Spl_1, Spl_Pattern)
        if hamming_dist <= d:
            count += 1
    return(count)
```

```python
#Function Neighbors(Pattern, d) generates the d-neighborhood that is the set of all k-mers whose Hamming distance
#from Pattern does not exceed d

def Neighbors(Pattern, d):
    nucleotides = ['A','C','G','T']
    if d == 0:
        return [Pattern]
    if len(Pattern) == 1:
        return nucleotides
    Neighborhood = []
    for neighbor in Neighbors(Pattern[1:],d):
        if HammingDistance(Pattern[1:],neighbor) < d:
            for nucl in nucleotides:
                Neighborhood.extend([nucl+neighbor])
        else:
            Neighborhood.append(Pattern[0] + neighbor)
    return Neighborhood
```

```python
#Function FrequentWordsWithMismatches(Text, k, d) finds the most frequent k-mers with mismatches in a string

def FrequentWordsWithMismatches(Text, k, d):
    Patterns = []
    freqMap = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:(k+i)]
        neighborhood = Neighbors(Pattern, d)
        for j in range(len(neighborhood)-1):
            neighbor = neighborhood[j]
            if neighbor not in freqMap.keys():
                freqMap[neighbor] = 1
            else:
                freqMap[neighbor] = freqMap[neighbor] + 1
    m = max(freqMap.values())
    for key_pat in freqMap:
        if freqMap[key_pat] == m:
            Patterns.append(key_pat)
    return(Patterns)
```

```python
#Function FreqWordsMismatch_and_ReverseCompl(Text, k, d) finds the most frequent k-mers (with mismatches and reverse complements) in a string

def FreqWordsMismatch_and_ReverseCompl(Text, k, d):
    Patterns = []
    freqMap = {}
    ReverseMap = {}
    SumMap = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:(k+i)]
        neighborhood = Neighbors(Pattern, d)
        for j in range(len(neighborhood)):
            neighbor = neighborhood[j]
            if neighbor not in freqMap.keys():
                freqMap[neighbor] = 1
            else:
                freqMap[neighbor] = freqMap[neighbor] + 1
        for l in range(len(neighborhood)):
            neighbor_1 = ReverseComplement(neighborhood[l])
            if neighbor_1 not in ReverseMap.keys():
                ReverseMap[neighbor_1] = 1
            else:
                ReverseMap[neighbor_1] = ReverseMap[neighbor_1] + 1

    for pat in freqMap.keys():
        if pat in ReverseMap.keys():
            SumMap[pat] = freqMap[pat] + ReverseMap[pat]
    m_sum = max(SumMap.values())
    for key_pat in SumMap.keys():
        if SumMap[key_pat] == m_sum:
            Patterns.append(key_pat)
    return Patterns
```

```python
#Function MotifEnumeration(Dna, k, d) finds all (k, d)-motifs in a collection of strings.

def MotifEnumeration(Dna, k, d):
    Patterns = []
    neighborhood = {}
    all_kmers = {}
    Dna_list = Dna.split(' ')
    for str_num in range(len(Dna_list)):
        all_kmers[str_num] = []
        for i in range(len(Dna_list[str_num])-k+1):
            all_kmers[str_num].append(Dna_list[str_num][i:i+k])

        for num in list(all_kmers.keys()):
            neighborhood[num] = []
            for k_mer in all_kmers[num]:
                neighborhood[num].extend(Neighbors(k_mer, d))
                neighborhood[num] = list(set(neighborhood[num]))

    count = {}
    for k in neighborhood.keys():
        for p in neighborhood[k]:
            count[p] = 0
    for key in neighborhood.keys():
        for Pattern_mism in count.keys():
            if Pattern_mism in neighborhood[key]:
                count[Pattern_mism] += 1
    
    for k_mism in count.keys():
        if count[k_mism] == len(Dna_list):
            Patterns.append(k_mism)
            
    return Patterns
```

```python
#Function all_kmers_generator(Dna, k) generates all possible k-mers of k-length nucleotides.

def all_kmers_generator(Dna, k):
    Dna_list = Dna.split(' ')
    all_kmers = {}
    for str_num in range(len(Dna_list)):
        all_kmers[Dna_list[str_num]] = []
        for i in range(len(Dna_list[str_num])-k+1):
            all_kmers[Dna_list[str_num]].append(Dna_list[str_num][i:i+k])
    return all_kmers
```

```python
#Function DistanceBetweenPatternAndStrings(Pattern, Dna) finds a sum of distances between Pattern and each string in Dna

def DistanceBetweenPatternAndStrings(Pattern, Dna):
    k = len(Pattern)
    distance = 0
    all_kmers = all_kmers_generator(Dna, k)   

    for str_pat in all_kmers.keys():
        HammingDist = float('inf')
        for k_mer_pat in all_kmers[str_pat]:
            if HammingDist > HammingDistance(Pattern, k_mer_pat):
                HammingDist = HammingDistance(Pattern, k_mer_pat)
        distance = distance + HammingDist
    return distance
```

```python
#Function MedianString(Dna, k) finds a median string (a k-mer Pattern that minimizes
# DistanceBetweenPatternAndStrings(Pattern, Dna) among all possible choices of k-mers)

import itertools
from itertools import product

def MedianString(Dna, k):
    distance = float('inf')

    perm_list = list(product(('A', 'T', 'C', 'G'), repeat=k))
    perm_list_1 = ['']*len(perm_list)
    for i in range(len(perm_list)):
        perm_list[i] = list(perm_list[i])
        for j in range(len(list(perm_list[i]))):
            perm_list_1[i] = perm_list_1[i] + perm_list[i][j]
    Patterns = perm_list_1

    for t in range(len(Patterns)):
        Pattern = Patterns[t]
        if distance > DistanceBetweenPatternAndStrings(Pattern, Dna):
            distance = DistanceBetweenPatternAndStrings(Pattern, Dna)
            Median = Pattern
    return Median
```

```python
#Function ProfileMostProbable_for_Greedy_Search(Text, k, Profile) is using for GreedyMotifSearch.

def NumberToSymbol(index):
    if index == 0:
        return str("A")
    elif index == 1:
        return str("C")
    elif index == 2:
        return str("G")
    elif index == 3:
        return str("T")

def window(s, k):
    for i in range(1 + len(s) - k):
        yield s[i:i+k]

def SymbolToNumber(Symbol):
    if Symbol == "A":
        return 0
    elif Symbol == "C":
        return 1
    elif Symbol == "G":
        return 2
    elif Symbol == "T":
        return 3


def ProfileMostProbable_for_Greedy_Search(Text, k, Profile):
    letter = [[] for key in range(k)]
    probable = ""
    hamdict = {}
    index = 1
    for a in range(k):
        for j in "ACGT":
            letter[a].append(Profile[j][a])
    for b in range(len(letter)):
        number = max(letter[b])
        probable += str(NumberToSymbol(letter[b].index(number)))
    for c in window(Text, k):
        for x in range(len(c)):
            y = SymbolToNumber(c[x])
            index *= float(letter[x][y])
        hamdict[c] = index
        index = 1
    for pat, ham in hamdict.items():
        if ham == max(hamdict.values()):
            final = pat
            break
            
    return final
```

```python
#Function Score(Motif_list) finds score of motifs.

import numpy as np
from typing import List

nuc_order = {'A': 0, 'C': 1,'G': 2,'T': 3}
def Profile(Motif_list: List[str], k: int) -> np.ndarray:
    Profile_1_dict = {'A': 0, 'C': 0,'G': 0,'T': 0}
    prob_mat = np.zeros((4, k))
    for nuc_seq in Motif_list:
        prob_mat += fill_prob_mat(nuc_seq, k)
    Profile_nump = prob_mat / np.sum(prob_mat, axis=0)
    for key_num in range(len(Profile_1_dict.keys())):
        Profile_1_dict[list(Profile_1_dict.keys())[key_num]] = list(Profile_nump[key_num, :])
    return Profile_1_dict

def fill_prob_mat(nuc_seq: str, k: int, nuc_order: dict = nuc_order) -> np.ndarray:
    prob_mat = np.zeros((4, k))
    for i, nuc in enumerate(nuc_seq):
        prob_mat[nuc_order[nuc], i] += 1
    return prob_mat

def Score(Motif_list):
    Sum = 0
    for nuc_ind in range(len(Motif_list[0])):
        nuc_col_dict = {}
        for seq in Motif_list:
            if seq[nuc_ind] not in nuc_col_dict.keys():
                nuc_col_dict[seq[nuc_ind]] = 1
            else:
                nuc_col_dict[seq[nuc_ind]] += 1
        count_col = len(Motif_list) - max(list(nuc_col_dict.values()))
        Sum += count_col
    return Sum
```

```python
#Function GreedyMotifSearch(Dna, k, t) tests to see whether Motifs outscores the current best scoring collection of motifs.

def GreedyMotifSearch(Dna, k, t):
    BestMotifs = []
    for string in Dna:
        BestMotifs.append(string[0:k])

    for nuc_i in range(len(Dna[0])-k+1):
        Motif_list = []
        Motif_list.append(Dna[0][nuc_i:nuc_i+k])

        for dna_i in range(1,t):
            Profile_1_dict = Profile(Motif_list[0:dna_i], k)
            Motif_list.append(ProfileMostProbable_for_Greedy_Search(Dna[dna_i], k, Profile_1_dict))
            
        if Score(Motif_list) < Score(BestMotifs):
            BestMotifs = Motif_list
            
    return BestMotifs
```

```python
# Greedy motif search with usage of pseudocounts (substitution zeroes with small numbers).

import numpy as np
from typing import List

nuc_order = {'A': 0, 'C': 1,'G': 2,'T': 3}
def Profile_with_pseudocounts(Motif_list: List[str], k: int) -> np.ndarray:
    Profile_1_dict = {'A': 0, 'C': 0,'G': 0,'T': 0}
    prob_mat = np.zeros((4, k))
    for nuc_seq in Motif_list:
        prob_mat += fill_prob_mat_with_pseudocounts(nuc_seq, k)
    Profile_nump = prob_mat / np.sum(prob_mat, axis=0)
    for key_num in range(len(Profile_1_dict.keys())):
        Profile_1_dict[list(Profile_1_dict.keys())[key_num]] = list(Profile_nump[key_num, :])
    return Profile_1_dict

def fill_prob_mat_with_pseudocounts(nuc_seq: str, k: int, nuc_order: dict = nuc_order) -> np.ndarray:
    prob_mat = np.zeros((4, k))
    for j in range(4):
        for g in range(k):
            prob_mat[j,g] = 1
    for i, nuc in enumerate(nuc_seq):
        prob_mat[nuc_order[nuc], i] += 1
    return prob_mat

def GreedyMotifSearch_with_pseudocounts(Dna, k, t):
    BestMotifs = []
    for string in Dna:
        BestMotifs.append(string[0:k])

    for nuc_i in range(len(Dna[0])-k+1):
        Motif_list = []
        Motif_list.append(Dna[0][nuc_i:nuc_i+k])

        for dna_i in range(1,t):
            Profile_1_dict = Profile_with_pseudocounts(Motif_list[0:dna_i], k)
            Motif_list.append(ProfileMostProbable_for_Greedy_Search(Dna[dna_i], k, Profile_1_dict))
            
        if Score(Motif_list) < Score(BestMotifs):
            BestMotifs = Motif_list
            
    return BestMotifs
```
